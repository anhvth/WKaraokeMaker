```python
from kmaker.data import *
DS = 'private'
def play_result(path):
    name = get_name(path)
    audio_path = glob(f'data/{DS}*/songs/{name}.*')[0]
    return ItemAudioLabel(path, audio_path)
```

```python
names = os.listdir(f'outputs/{DS}_with_masked_text/submission/')
name = names[6]
it = play_result(f'./outputs/{DS}_with_masked_text/submission/{name}')
play_audio(it.audio)
it.play('word')
```
